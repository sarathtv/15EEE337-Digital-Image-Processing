```python
import cv2
from skimage.filters import median
from google.colab.patches import cv2_imshow
```

```python

#Needs 8 bit, not float.
img_gaussian_noise = cv2.imread('/content/drive/MyDrive/images/Osteosarcoma_01_25Sigma_noise.tif', 0)
img_salt_pepper_noise = cv2.imread('/content/drive/MyDrive/images//Osteosarcoma_01_8bit_salt_pepper.tif', 0)

img = img_salt_pepper_noise


median_using_cv2 = cv2.medianBlur(img, 3)

from skimage.morphology import disk  
#Disk creates a circular structuring element, similar to a mask with specific radius
median_using_skimage = median(img, disk(3), mode='constant', cval=0.0)


cv2_imshow(img)
```
